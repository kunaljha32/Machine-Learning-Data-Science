```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://www.bankbazaar.com/reviews.html'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
```

```python
review_text=[]
review_text_elem=soup.find_all(class_='text_here review-desc-more')
#review_text_elem
for item in review_text_elem:
    review_text.append(item.text)

res = [sub.replace('\n', '') for sub in review_text] 

```

```python
user_name=[]
user_name_elem=soup.find_all(class_='js-author-name')
for item in user_name_elem:
    user_name.append(item.text)
#print(user_name)    
```

```python
bank_name=[]
bank_name_elem=soup.find_all('div',attrs={'class':'review-bank-title'})
for item in bank_name_elem:
    bank_name.append(item.find('img').get('alt'))
#print(bank_name)    
#bank_name
```

```python
final=[]
for text,user,bank in zip(res,user_name,bank_name):
    final.append({'Review_text':text,'User':user,'Bank':bank})
df=pd.DataFrame(final)
df
```
