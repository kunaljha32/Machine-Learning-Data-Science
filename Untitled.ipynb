```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('task_data.csv')
df.set_index("sample index", inplace = True)
df.head(10)
```

```python
df['class_label'].replace(to_replace =-1, value =np.nan,inplace=True)
```

```python

df.tail()
```

```python
df['class_label'].fillna(0, inplace=True)
df.tail()
```

```python
df.isnull().any()
```

```python
df.info()
```

```python

X = df.iloc[:,1:]
X
```

```python
Y = df['class_label']
Y
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.SelectKBest.html
#applying SelectKBest class to get scores of 10 features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit=bestfeatures.fit(X,Y)
```

```python

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
#concat two dataframes  
sensorScores = pd.concat([dfcolumns,dfscores],axis=1)


#naming the dataframe columns
sensorScores.columns = ['Sensors','Score'] 
```

```python
sensorScores
```

```python
print(sensorScores.nlargest(10,'Score')) 
```

```python
#Using feature importance of Extra tree classifier
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X,Y)
```

```python
print(model.feature_importances_) 
```

```python
#plot graph of feature importances for visualization
feature_importances = pd.Series(model.feature_importances_, index=X.columns)
feature_importances.nlargest(10).plot(kind='bar')
plt.title("Feature importances")
plt.show()
```
